```python
import numpy as np

# Training data for AND function
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

Y = np.array([0, 0, 0, 1])   # Desired output (AND)

# Initial parameters (given in question)
w1 = 0.3
w2 = -0.2
alpha = 0.2      # Learning rate
theta = 0.4      # Bias (threshold)

# Step activation function
def step_function(x):
    if x >= 0:
        return 1
    else:
        return 0

epoch = 0
converged = False

while not converged:
    epoch += 1
    print(f"\nEpoch {epoch}")
    converged = True

    for i in range(len(X)):
        x1, x2 = X[i]

        # Net input
        weighted_sum = (x1 * w1 + x2 * w2) - theta

        # Predicted output
        y_pred = step_function(weighted_sum)

        # Error
        error = Y[i] - y_pred

        print(f"Input: ({x1},{x2})  Weighted Sum: {weighted_sum:.2f}  "
              f"Y_pred: {y_pred}  Error: {error}  "
              f"W1: {w1:.2f}  W2: {w2:.2f}")

        # Update weights if error exists
        if error != 0:
            w1 = w1 + alpha * error * x1
            w2 = w2 + alpha * error * x2
            converged = False

print("\nTraining completed!")
print(f"Final Weights: W1 = {w1}, W2 = {w2}")
```
